```python
target_cls = ['46','47','49']
```

```python
map_cls = dict()
map_cls['46'] = '0'
map_cls['47'] = '1'
map_cls['49'] = '2'
```

```python
labels = glob.glob('./coco/labels/val/*')
```

```python
len(labels)
```

```python
for item in labels :
    with open(item,'r') as file :
        data = file.readlines()
    with open(item,'w') as file :
        for line in data : 
            lines = line.split()
            if lines[0] in target_cls :
                lines[0] = map_cls[lines[0]]
                l = ' '.join(lines) + '\n'
                file.write(l)
```
